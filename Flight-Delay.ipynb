```python
ori_flights['SCHEDULED_DEP_HOUR'] = ori_flights['SCHEDULED_DEPARTURE'].str[:2]
ori_flights['SCHEDULED_DEP_MIN'] = ori_flights['SCHEDULED_DEPARTURE'].str[2:]

ori_flights['DEP_HOUR'] = ori_flights['DEPARTURE_TIME'].str[:2]
ori_flights['DEP_MIN'] = ori_flights['DEPARTURE_TIME'].str[2:]

ori_flights['WHEELS_OFF_HOUR'] = ori_flights['WHEELS_OFF'].str[:2]
ori_flights['WHEELS_OFF_MIN'] = ori_flights['WHEELS_OFF'].str[2:]

ori_flights['WHEELS_ON_HOUR'] = ori_flights['WHEELS_ON'].str[:2]
ori_flights['WHEELS_ON_MIN'] = ori_flights['WHEELS_ON'].str[2:]

ori_flights['SCHEDULED_ARRIVAL_HOUR'] = ori_flights['SCHEDULED_ARRIVAL'].str[:2]
ori_flights['SCHEDULED_ARRIVAL_MIN'] = ori_flights['SCHEDULED_ARRIVAL'].str[2:]

ori_flights['ARRIVAL_TIME_HOUR'] = ori_flights['ARRIVAL_TIME'].str[:2]
ori_flights['ARRIVAL_TIME_MIN'] = ori_flights['ARRIVAL_TIME'].str[2:]
```

### Let's Remove all the cancelled flights information as our focus is predicting Delay. 

```python
ori_flights = ori_flights[['YEAR','MONTH','DAY','DAY_OF_WEEK','AIRLINE','FLIGHT_NUMBER','TAIL_NUMBER','ORIGIN_AIRPORT','DESTINATION_AIRPORT','SCHEDULED_DEP_HOUR','SCHEDULED_DEP_MIN','DEP_HOUR','DEP_MIN','DEPARTURE_DELAY','TAXI_OUT','WHEELS_OFF_HOUR','WHEELS_OFF_MIN','SCHEDULED_TIME','ELAPSED_TIME','AIR_TIME','DISTANCE','WHEELS_ON_HOUR','WHEELS_ON_MIN','TAXI_IN','SCHEDULED_ARRIVAL_HOUR','SCHEDULED_ARRIVAL_MIN','ARRIVAL_TIME_HOUR','ARRIVAL_TIME_MIN','ARRIVAL_DELAY','DIVERTED','CANCELLED','CANCELLATION_REASON','AIR_SYSTEM_DELAY','SECURITY_DELAY','AIRLINE_DELAY','LATE_AIRCRAFT_DELAY','WEATHER_DELAY']]
```

```python
ori_flights.head()
```

```python
flights = ori_flights[ori_flights['CANCELLED'] == '0']
flights = ori_flights[ori_flights['DIVERTED'] == '0']
flights.drop(['CANCELLATION_REASON','AIR_SYSTEM_DELAY','SECURITY_DELAY','AIRLINE_DELAY','LATE_AIRCRAFT_DELAY','WEATHER_DELAY'],axis=1,inplace=True)
```
